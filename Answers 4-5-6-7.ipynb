## Answer to question 4

```python
# Create a dictionary to store minimum daily vaccination number for each country
min_daily_vaccinations = {}

# Iterate over each country in the dataset
for country in df["country"].unique():
    
    # Get the minimum daily vaccination number for the country
    min_daily_vacc = df[df["country"] == country]["daily_vaccinations"].min()
    
    # Store the minimum daily vaccination number in the dictionary
    min_daily_vaccinations[country] = min_daily_vacc

# Fill missing values in daily_vaccinations column with the minimum daily vaccination number of relevant countries
df["daily_vaccinations"] = df.groupby("country")["daily_vaccinations"].apply(lambda x: x.fillna(min_daily_vaccinations[x.name] if not np.isnan(min_daily_vaccinations[x.name]) else 0))
```

```python
df.head(20)
```

## Answer to question 5

```python
# Calculate the median daily vaccinations per country
median_daily_vaccinations = df.groupby("country")["daily_vaccinations"].median()

# Sort the countries by median daily vaccinations in descending order
sorted_countries = median_daily_vaccinations.sort_values(ascending=False)

# Get the top 3 countries with highest median daily vaccinations
top_3_countries = sorted_countries.head(3)

# Print the top 3 countries
print(top_3_countries)
```

## Answer to question 6

```python
# filter the dataframe to only include rows where date is 1/6/2021
df_filtered = df[df["date"] == "1/6/2021"]

# sum the daily_vaccinations column for the filtered dataframe
total_vaccinations = df_filtered["daily_vaccinations"].sum()

print("Total vaccinations on 1/6/2021:", total_vaccinations)
```
